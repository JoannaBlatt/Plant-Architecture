```python
import pandas as pd
```

```python
cyto_d5_df = pd.read_csv('pimpi_Cyto_D5_Root_Nodes.csv', skipinitialspace=True)
cyto_d5_df
```

There is one image where the plate number was marked as p1#3. We are not sure whether this should have been part of pl3 or something else, so we simply deleted this image

```python
cyto_d5_df.drop(cyto_d5_df.loc[cyto_d5_df['image'] == '20221119_RSA_M248M058LA1511_cyto_100mM_Salt050.rsml'].index, axis=0, inplace=True)
cyto_d5_df
```

```python
def reformat_root_name(root_name):
    root_name_items = root_name.split('_')
    if len(root_name_items) == 2:
        return root_name
    
    plate, condition, hormone, genotype = root_name.split('_')
    condition = condition.capitalize()
    replicate = plate.strip('pl')
    if not replicate.isnumeric():
        print(replicate)
    assert replicate.isnumeric()
    return '%s_%s_%s_%s' % (genotype, replicate, condition, hormone)
```

```python
cyto_d5_df['root_name'] = cyto_d5_df['root_name'].apply(reformat_root_name)
cyto_d5_df
```

```python
cyto_d5_df['image'] = '_set1_day5_' + cyto_d5_df['image']
cyto_d5_df
```

```python
cyto_d5_df.to_csv('pimpi_Cyto_D5_Root_Nodes.csv', index=False)
```
