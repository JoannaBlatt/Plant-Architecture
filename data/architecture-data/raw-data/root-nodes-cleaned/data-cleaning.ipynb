```python
import pandas as pd
```

```python
df = pd.read_csv('pimpi_IAA1_D5_Root_Nodes.csv', skipinitialspace=True)
df
```

```python
def reformat_root_name(root_name):
    root_name_items = root_name.split('_')
    if len(root_name_items) == 2:
        return root_name
    else:
        plate, condition, hormone, genotype = root_name_items
        replicate = plate.strip('pl')
        condition = condition.capitalize()
        return '%s_%s_%s_%s' % (genotype, replicate, condition, hormone)
```

```python
df['root_name'] = df['root_name'].apply(reformat_root_name)
```

```python
df.to_csv('pimpi_IAA1_D5_Root_Nodes.csv', index=False)
```

```python
df = pd.read_csv('pimpi_IAA1_D9_Root_Nodes.csv', skipinitialspace=True)
df['root_name'] = df['root_name'].apply(reformat_root_name)
df.to_csv('pimpi_IAA1_D9_Root_Nodes.csv', index=False)
```
